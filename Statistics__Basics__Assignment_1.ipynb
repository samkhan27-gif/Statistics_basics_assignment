### Ques 2 :  What is sampling in statistics? Explain the differences between random and stratified sampling.

#### Ans : Sampling is the process of selecting a small group (sample) from a large group (population) in order to study it. Studying the whole population is often difficult, expensive, or time-consuming, so we use a sample to understand the population.

> in simple words, sampling helps us save time and effort while still getting useful information.

**Example**:

> If a city has 1,00,000 voters, it is not practical to ask everyone about their opinion. Instead, a researcher may select 1,000 voters as a sample and study their responses.

> ## Random Sampling

> Random sampling means every member of the population has an equal chance of being selected. The selection is completely by chance, with no preference.
Key idea: Fair and unbiased selection.

**Example**:

- A teacher wants to know how students feel about online classes. She writes all students names on slips, mixes them in a box, and picks 30 names randomly.
 Each student had the same chance of being selected.

>>  **Advantages** :

>>- Simple to use


>>- Reduces personal bias


>>- Fair representation (by chance)


> ## Stratified Sampling

> Stratified sampling divides the population into groups (called strata) based on a specific characteristic (such as gender, age, class, or income). Then, random samples are taken from each group.
Key idea: Representation of all important groups.

**Example** : A school has:

>- 60% boys


>- 40% girls


> To study student performance, the researcher :

>>- Separates students into boys and girls

>>- Randomly selects students from each group in the same proportion

>> This ensures both boys and girls are properly represented.

>> **Advantages** :

>>- More accurate results


>>- Ensures representation of all groups



### Ques 3 : Define mean, median, and mode. Explain why these measures of central tendency are important.

#### Ans :   Mean
 Mean is the average of a set of values. It is calculated by adding all the values and dividing the sum by the total number of values.

**Formula** :

 **Mean** = (Sum of all observations) / (Number of observations)

 **Example** :

> If the marks are 10, 20, 30, 40, 50
 Mean = (10 + 20 + 30 + 40 + 50) / 5 = 30

 **Median**

 Median is the middle value of a dataset when the values are arranged in ascending or descending order.

**Example** :

- Data: 10, 20, 30, 40, 50

- Median = 30

> If there are even numbers of observations, the median is the average of the two middle values.

> **Mode**

 Mode is the value that occurs most frequently in a dataset.

**Example** :

 > Data: 2, 4, 4, 6, 8

> Mode = 4 (because it appears the most times)



>> ### Importance of Measures of Central Tendency : ###

> Measures of central tendency are important because they help us understand the central or typical value of a dataset.

>- They summarize large data into a single value


>- They make comparison between different datasets easy


>- Mean is useful for overall analysis

>- Median is useful when data has extreme values


>- Mode is useful to find the most common value


> These measures are widely used in statistics, economics, education, and data analysis for decision making.






### Ques 4 : Explain skewness and kurtosis. What does a positive skew imply about the
Data?
#### Ans :  Skewness

Skewness is a measure that shows the asymmetry of a data distribution around its mean. It tells us whether the data is more spread on the left side or the right side.


> **Types of Skewness** :

>- Positive Skewness (Right Skewed): Long tail on the right side


>- Negative Skewness (Left Skewed): Long tail on the left side


>- Zero Skewness: Symmetrical distribution


>> ##  Kurtosis

 > Kurtosis is a measure that describes the peakedness or flatness of a data distribution compared to a normal distribution. It indicates how heavy or light the tails of the distribution are.

> **Types of Kurtosis** :

>- Leptokurtic: High peak, heavy tails


>- Mesokurtic: Normal peak


>- Platykurtic: Flat peak, light tails

> **What Does a Positive Skew Imply About the Data?**

>A positive skew implies that:

>> - The distribution has a long tail on the right side


>> - Most data values are concentrated on the lower side


>> - A few extremely high values pull the mean towards the right


>> Generally: **Mean > Median > Mode**


**Example**:

> Income distribution is usually positively skewed because most people earn moderate income, while a few people earn very high income.





### Ques 5 : Implement a Python program to compute the mean, median, and mode of
a given list of numbers.
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]


#### Ans : Python Program to Calculate Mean, Median, and Mode




```python
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

# Mean
mean = sum(numbers) / len(numbers)

# Median
numbers_sorted = sorted(numbers)
n = len(numbers_sorted)

if n % 2 == 0:
    median = (numbers_sorted[n//2 - 1] + numbers_sorted[n//2]) / 2
else:
    median = numbers_sorted[n//2]

# Mode
frequency = {}
for num in numbers:
    frequency[num] = frequency.get(num, 0) + 1

mode = max(frequency, key=frequency.get)

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
```

### Ques 6 : Compute the covariance and correlation coefficient between the following
two datasets provided as lists in Python:
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]


#### Ans : Covariance shows the direction of relationship between two variables.

>- Correlation coefficient measures the strength and direction of the relationship.

>- The high positive correlation value indicates a strong positive relationship between the datasets.

```python
import math

list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

n = len(list_x)

mean_x = sum(list_x) / n
mean_y = sum(list_y) / n

# Covariance
covariance = sum((list_x[i] - mean_x) * (list_y[i] - mean_y) for i in range(n)) / n

# Standard deviations
std_x = math.sqrt(sum((x - mean_x) ** 2 for x in list_x) / n)
std_y = math.sqrt(sum((y - mean_y) ** 2 for y in list_y) / n)

# Correlation coefficient
correlation = covariance / (std_x * std_y)

print("Covariance:", covariance)
print("Correlation Coefficient:", correlation)
```

### Ques 7 :  Write a Python script to draw a boxplot for the following numeric list and identify its outliers. Explain the result:
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

#### Ans : A boxplot visually represents the distribution of data using quartiles and highlights extreme values. In this dataset, 35 lies beyond the upper bound (Q3 + 1.5 * IQR), so it is identified as an outlier.

> This indicates that most values are concentrated between 12 and 29, while 35 is unusually high compared to the rest.

```python
import matplotlib.pyplot as plt
import numpy as np

data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

# Draw boxplot
plt.boxplot(data)
plt.title("Boxplot of Given Data")
plt.ylabel("Values")
plt.show()

# Calculate quartiles
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

# Identify outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = [x for x in data if x < lower_bound or x > upper_bound]

print("Outliers:", outliers)
```

#### Ques 8 : You are working as a data analyst in an e-commerce company. The marketing team wants to know if there is a relationship between advertising spend and daily sales.

● Explain how you would use covariance and correlation to explore this
relationship.

● Write Python code to compute the correlation between the two lists:
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]


#### Ans : To understand the relationship between advertising spend and daily sales, I would use covariance and correlation.

> - Covariance helps identify the direction of the relationship between two variables.

> If covariance is positive, it means that as advertising spend increases, daily sales also tend to increase.

> If it is negative, one increases while the other decreases.

>- Correlation coefficient measures the strength and direction of the relationship.
Its value lies between -1 and +1.
A value close to +1 indicates a strong positive relationship, while a value close to 0 indicates little or no relationship.

> By using both covariance and correlation, we can understand whether increased advertising spend is associated with higher sales and how strong that relationship is.

```python
import math

advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

n = len(advertising_spend)

mean_x = sum(advertising_spend) / n
mean_y = sum(daily_sales) / n

covariance = sum(
    (advertising_spend[i] - mean_x) * (daily_sales[i] - mean_y)
    for i in range(n)
) / n

std_x = math.sqrt(sum((x - mean_x) ** 2 for x in advertising_spend) / n)
std_y = math.sqrt(sum((y - mean_y) ** 2 for y in daily_sales) / n)

correlation = covariance / (std_x * std_y)

print("Correlation Coefficient:", correlation)
```

### Ques 9 : Your team has collected customer satisfaction survey data on a scale of 1-10 and wants to understand its distribution before launching a new product.
● Explain which summary statistics and visualizations (e.g. mean, standard
deviation, histogram) you'd use.

● Write Python code to create a histogram using Matplotlib for the survey data:
survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]


#### Ans : To understand the distribution of customer satisfaction scores before launching a new product, I would use the following summary statistics and visualizations:

**Mean** :

It shows the average satisfaction level of customers.


**Median** :

It helps identify the middle value and is useful if the data has extreme values.


**Standard Deviation** :

> It measures how much the satisfaction scores vary from the average. A low value means responses are consistent, while a high value indicates mixed opinions.


**Histogram** :

> A histogram visually shows how frequently each score occurs. It helps identify patterns such as whether customers are mostly satisfied, dissatisfied, or evenly distributed.

>> Using these measures together gives a clear picture of overall satisfaction and variation in customer responses.


```python
import matplotlib.pyplot as plt

survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

plt.hist(survey_scores, bins=7)
plt.xlabel("Survey Scores")
plt.ylabel("Frequency")
plt.title("Histogram of Customer Satisfaction Scores")
plt.show()
```
